```python
def get_wx(xo, xw, xf, xdoto, xdotw, xdotf, to, tw, tf):
    
    Btw = np.asarray(get_bernstein_poly(to,tw,tf))
    Bdottw = np.asarray(get_bernstein_differentials(to,tw,tf))
    
    Bto = np.asarray(get_bernstein_poly(to,to,tf))
    Bdotto = np.asarray(get_bernstein_differentials(to,to,tf))
    
    Btf = np.asarray(get_bernstein_poly(to,tf,tf))
    Bdottf = np.asarray(get_bernstein_differentials(to,tf,tf))

    B = np.array([[Btw[1], Btw[2], Btw[3], Btw[4]],
                 [Bdotto[1], Bdotto[2], Bdotto[3], Bdotto[4]],
                 [Bdottf[1], Bdottf[2], Bdottf[3], Bdottf[4]],
                 [Bdottw[1], Bdottw[2], Bdottw[3], Bdottw[4]]])
    
    A = np.array([[xw - xo*Btw[0] - xf*Btw[5]],
                 [xdoto - xo*Bdotto[0] - xf*Bdotto[5]],
                 [xdotf - xo*Bdottf[0] - xf*Bdottf[5]],
                 [xdotw - xo*Bdottw[0] - xf*Bdottw[5]]])
    
    #In get coefficient function these coefficients are taken as x0,x1,x2,x3,x4,x5
    W = np.linalg.pinv((B.astype(float)))@(A.astype(float))
    W0 = xo
    Wf = xf
    W1 = W[0,0]
    W2 = W[1,0]
    W3 = W[2,0]
    W4 = W[3,0]
    
    return W0,W1,W2,W3,W4,Wf
```

```python
#3
def get_wk(yo, yw, yf, ko, kf, kdoto, kdotf, xo, xw, xf, xdoto, xdotw, xdotf, to,tw,tf):
    
    Btw = np.asarray(get_bernstein_poly(to,tw,tf))
    Bdottw = np.asarray(get_bernstein_differentials(to,tw,tf))
    
    Bto = np.asarray(get_bernstein_poly(to,to,tf))
    Bdotto = np.asarray(get_bernstein_differentials(to,to,tf))
    
    Btf = np.asarray(get_bernstein_poly(to,tf,tf))
    Bdottf = np.asarray(get_bernstein_differentials(to,tf,tf))
    
    x0,x1,x2,x3,x4,x5 = get_wx(xo, xw, xf, xdoto, xdotw, xdotf, to, tw, tf)
    
    coefftw = np.asarray(get_product_functions( x0,x1,x2,x3,x4,x5,to,tw,tf ))
    coefftf = np.asarray(get_product_functions( x0,x1,x2,x3,x4,x5,to,tf,tf ))
    coeffto = np.asarray(get_product_functions( x0,x1,x2,x3,x4,x5,to,to,tf ))
    
    B = np.array([[Bdotto[1], Bdotto[2], Bdotto[3], Bdotto[4]],
                 [Bdottf[1], Bdottf[2], Bdottf[3], Bdottf[4]],
                 [coefftf[1], coefftf[2], coefftf[3], coefftf[4]],
                 [coefftw[1], coefftw[2], coefftw[3], coefftw[4]],
                 [coeffto[1], coeffto[2], coeffto[3], coeffto[4]]])
    
    A = np.array([[kdoto - ko*Bdotto[0] - kf*Bdotto[5]],
                 [kdotf - ko*Bdottf[0] - kf*Bdottf[5]],
                 [yf - ko*coefftf[0] - kf*coefftf[5] - yo],
                 [yw - ko*coefftw[0] - kf*coefftw[5] - yo],
                 [yo - ko*coeffto[0] - kf*coeffto[5] - yo]])
    
    W = np.linalg.pinv((B.astype(float)))@(A.astype(float))
    
    k0 = ko
    k5 = kf
    k1 = W[0,0]
    k2 = W[1,0]
    k3 = W[2,0]
    k4 = W[3,0]
    
    return k0,k1,k2,k3,k4,k5
```

```python
class agent:
    def __init__(self,rad,tf,tw,start,waypt,goal,init_or,final_or,xdotw, obxo, obyo):
        
        self.init_or = init_or #initial orientation
        self.final_or = final_or #final orientation
        self.xdoto=0 #initial velocity 
        self.xdotf=0 #final velocity
        
        self.xdotw= xdotw #waypoint velocity #I ADDED
        
        self.k0=np.tan(self.init_or)   # tangent of the initial heading angle of the robot
        self.kdot0=0 # differntial of the tangent of the initial heading angle of the robot
        self.kdotw = 0# differntial of the tangent of the waypoint heading angle of the robot
        self.kf=np.tan(self.final_or)     #  tangent of the final heading angle of the robot
        self.kdotf=0  # differntial of the tangent of the final heading angle of the robot
        self.rad = rad # Radius of Robot
        self.t0 = 0 #time start
        self.tf = tf #time final
        self.tw = tw #time waypoint
        self.pos = copy.deepcopy(start) #holds current position [xp, yp]
        self.x = [] #holds x values
        self.y = [] #holds y values
        self.xdot = [] #holds x velocities
        self.ydot = [] #holds y velocities
        self.goal = goal #[xg, yg]
        self.origin = start #[xs, ys]
        self.waypt = waypt #[xw, yw]
        self.xf = self.goal[0] #final x
        self.yf = self.goal[1] #final y
        
        self.obxo = obxo
        self.obyo = obyo
        self.obx = [obxo]
        self.oby = [obyo]
        
        self.dt = 0.1
            
    def get_overall_traj(self):
        self.obstacle_traj_gen()
        
        x0, x1, x2, x3, x4, x5 = get_wx(self.origin[0], self.waypt[0], self.goal[0], self.xdoto, self.xdotw, self.xdotf, self.t0, self.tw, self.tf)
        k0, k1, k2, k3, k4, k5 = get_wk(self.origin[1], self.waypt[1], self.goal[1], self.k0, self.kf, self.kdot0, self.kdotf, self.origin[0], self.waypt[0], self.goal[0], self.xdoto, self.xdotw, self.xdotf, self.t0, self.tw, self.tf)
        
        for t in np.arange(self.t0,self.tf + self.dt,self.dt):
            t = round(t,1)            
            B = np.asarray(get_bernstein_poly(self.t0,t,self.tf))
            Bdot = np.asarray(get_bernstein_differentials(self.t0,t,self.tf))
            coeff = np.asarray(get_product_functions( x0,x1,x2,x3,x4,x5,self.t0,t,self.tf ))
            
            x = B[0]*x0 + B[1]*x1 + B[2]*x2 + B[3]*x3 + B[4]*x4 + B[5]*x5
            xdot = Bdot[0]*x0 + Bdot[1]*x1 + Bdot[2]*x2 + Bdot[3]*x3 + Bdot[4]*x4 + Bdot[5]*x5
            k = B[0]*k0 + B[1]*k1 + B[2]*k2 + B[3]*k3 + B[4]*k4 + B[5]*k5
            kdot = Bdot[0]*k0 + Bdot[1]*k1 + Bdot[2]*k2 + Bdot[3]*k3 + Bdot[4]*k4 + Bdot[5]*k5
            y = self.origin[1] + k0 * coeff[0] + k1 * coeff[1] + k2 * coeff[2] + k3 * coeff[3] + k4 * coeff[4] + k5 * coeff[5]
            ydot = xdot*k
            
            self.x.append(x)
            self.y.append(y)
            self.xdot.append(xdot)
            self.ydot.append(ydot)
            self.pos[0] = x
            self.pos[1] = y-1
            self.visualize_traj(t)
        
    def obstacle_traj_gen(self):
        #SIMPLE OBSTACLE USED FOR ALL SUBPARTS. 
        xlen = (self.waypt[0] - self.obxo)
        ylen = (self.waypt[1] - self.obyo)
        tlen = int(round(((self.tw - self.t0)/self.dt),0))
        dx = xlen/tlen
        dy = ylen/tlen
        
        for t in np.arange(self.t0 + self.dt,self.tf + self.dt,self.dt):
            self.obx.append(self.obx[-1] + dx)
            self.oby.append(self.oby[-1] + dy)
            
    def visualize_traj(self,time):
        """
        Function to visualize robot trajectory at a given timestamp. 
        Saves snapshots to /data folder as snap0.png, snap1.png...
        String the snapshots using the bash script mkmovie.sh to generate simulation video.
        """
        figure = plt.figure()
        ax = figure.add_subplot(1,1,1)
        obstacle = matplotlib.patches.Circle(
            (self.obx[len(self.x)-1],self.oby[len(self.x)-1]),
            radius = self.rad,
            facecolor ='red',
            linewidth=1.0,
            ls='solid',
            alpha=1,
            zorder=2,
            edgecolor = 'black')
        ax.add_patch(obstacle)
        
#         Sobstacle = matplotlib.patches.Circle(
#             (5,5),
#             radius = self.rad,
#             facecolor ='red',
#             linewidth=1.0,
#             ls='solid',
#             alpha=1,
#             zorder=2,
#             edgecolor = 'black')
#         ax.add_patch(Sobstacle)
        
        robot = matplotlib.patches.Circle(
            (self.x[len(self.x)-1],self.y[len(self.x)-1]),
            radius = self.rad,
            edgecolor='black',
            linewidth=1.0,
            ls='solid',
            alpha=1,
            zorder=2)
        ax.add_patch(robot)
        name = 'data/snap%d.png'%int(time*10)
        ax.plot([self.goal[0]], [self.goal[1]], '*', color="red", markersize =15,linewidth=3.0)
        ax.plot([self.waypt[0]], [self.waypt[1]], '*', color="green", markersize =15,linewidth=3.0)
        ax.plot(self.x,self.y,"b--")
        ax.plot(self.obx[:len(self.x)-1],self.oby[:len(self.y)-1],"r--")
        
        ax.set_aspect('equal')
        ax.set_xlim((-5,25))
        ax.set_ylim((-5,25))
        ax.set_xlabel(r'$x (m)$')
        ax.set_ylabel(r'$y (m)$')
        ax.grid(True)
        plt.savefig(name, dpi = 200)
        plt.cla()
        plt.close(figure)
        return figure
```

```python
#finding the current max velocity
mx = 0
mxi = 0
for i in range(len(bot.x)-1):
    dx = bot.x[i+1] - bot.x[i]
    dy = bot.y[i+1] - bot.y[i]
    if(mx != max(mx, (dx/bot.dt)**2 + (dy/bot.dt)**2)):
        mxi = i
    mx = max(mx, (dx/bot.dt)**2 + (dy/bot.dt)**2)
    
vmax = math.sqrt(mx)
Vlimit = 10
#choosing constant k so that it goes to Vlimit
k = Vlimit/vmax
print(f'Chosen value of constant k : {k}')
```

```python
#store position, velocity, time
robot_before_info = []
for i in range(len(bot.x)):
    if(i == len(bot.x) - 1):
        velx = vely = 0
    else:
        velx = (bot.x[i+1] - bot.x[i])/bot.dt
        vely = (bot.y[i+1] - bot.y[i])/bot.dt
    
    infoi = [(bot.x[i],bot.y[i]),(velx,vely),i*bot.dt]
    robot_before_info.append(infoi)
```

```python
def check_sensor(rx,ry,ox,oy):
    #sensor senses when centre of object is 5m away from robot
    if((rx-ox)**2 + (ry-oy)**2 < 5**2):
        return True
    else:
        return False

#note that this will change for change in dt
def check_collide(ri,bot):
    t = ri[2]
    to = math.floor(t*10)
    ox = bot.obx[to] + (bot.obx[to+1] - bot.obx[to])*(t-to/10)/bot.dt
    oy = bot.oby[to] + (bot.oby[to+1] - bot.oby[to])*(t-to/10)/bot.dt
    if(check_sensor(*ri[0],ox,oy)):
        return True
    else:
        return False
```

```python
def finds(quad,s):
    if(quad.subs([(s,1)]) >= 0):
        return 1
    
    end = 2.52
    start = 1
    
    while(abs(start-end)<=0.01):
        mid = round((start+end)/2,2)
        if(quad.subs([(s,mid)]) < 0):
            start = mid
        elif(quad.subs([(s,mid)]) > 0):
            end = mid
        else:
            return min(mid+0.1,2.52)
    return end
```

## Time scaling

```python
def check_sensor(rx,ry,ox,oy):
    #sensor senses when centre of object is 5m away from robot
    if((rx-ox)**2 + (ry-oy)**2 < 5**2):
        return True
    else:
        return False

    
def check_collide(ri,r2i):
    if(check_sensor(ri[0][0],ri[0][1],r2i[0][0],r2i[0][1])):
        return True
    else:
        return False
```

```python
robot_after_info = []
robot2_after_info = []
k = 3

i = 0
while i < len(bot.x):
    robot_after_info.append(copy.deepcopy(robot_before_info[i]))
    robot2_after_info.append(copy.deepcopy(robot2_before_info[i]))
    
    posx,posy = copy.deepcopy(robot_after_info[i][0])
    pos2x,pos2y = copy.deepcopy(robot2_after_info[i][0])
    
    
    if(check_sensor(posx,posy,pos2x,pos2y)):
        break
    i+=1

j = i
       
while(check_collide(robot_after_info[i],robot2_after_info[j])):
    velx = copy.deepcopy(robot_before_info[i][1][0])
    vely = copy.deepcopy(robot_before_info[i][1][1])
    vel2x = copy.deepcopy(robot2_before_info[j][1][0])
    vel2y = copy.deepcopy(robot2_before_info[i][1][1])
    
    robot_after_info[i][1] = (k*velx,k*vely)
    vx2 = (1/k)*vel2x
    vy2 = (1/k)*vel2y
#     print(j,i)
    robot2_after_info[j][1] = (vx2,vy2)
       
    i+=1
    j+=1
       
    pos2x,pos2y = copy.deepcopy(robot2_before_info[j][0])
    t2 = robot_after_info[j-1][2] + bot.dt*k
    
    #chosen k such that, if triggered, first robot will be 16x faster than 2nd
    for l in range(9):
        posx,posy = copy.deepcopy(robot_before_info[i][0])       
        t = robot_after_info[i-1][2] + bot.dt/k
        infoi = [(posx,posy), (0,0), t]
        
        robot_after_info.append(infoi)
        
        velx,vely = copy.deepcopy(robot_before_info[i][1][0]),copy.deepcopy(robot_before_info[i][1][1])
        robot_after_info[i][1] = (k*velx,k*vely)
        if(l!=3):
            i+=1

    info2i = [(pos2x,pos2y), (0,0), t2]
       
    robot2_after_info.append(info2i)
       
    

while(len(robot_after_info)<len(robot_before_info)):
    velx,vely = copy.deepcopy(robot_before_info[i][1][0]),copy.deepcopy(robot_before_info[i][1][1])
    robot_after_info[i][1] = (velx,vely)
    i+=1
    
    posx,posy = copy.deepcopy(robot_before_info[i][0])
    t = robot_after_info[i-1][2] + bot.dt
    infoi = [(posx,posy), (0,0), t]
    robot_after_info.append(infoi)

while(len(robot2_after_info)<len(robot2_before_info)):
    vel2x,vel2y = copy.deepcopy(robot2_before_info[i][1][0]),copy.deepcopy(robot2_before_info[i][1][1])
    robot2_after_info[j][1] = (vel2x,vel2y)
    j+=1
    
    pos2x,pos2y = copy.deepcopy(robot2_before_info[j][0])
    t2 = robot2_after_info[j-1][2] + bot.dt
    info2i = [(pos2x,pos2y), (0,0), t2]
    robot2_after_info.append(info2i)
```
